### Logistic regression to predict the absenteeism.

Classification of people from dataset into 'moderately absent' and 'excessively absent' based on whether below or above the median from column 'Absenteeism Time in Hours'. 

Median is chosen as it is a numerically stable method, as it implicitly balances the data into roughly two halfs and minimises the ability of the machine to focus exclusively on only one of the categories (i.e. either 1 or 0). 

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
# Creating Targets using 0 & 1 values for 'Absenteeism Time in Hours' column. 
# Excessively absent - 1, Moderately absent - 0.
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] >
                   data_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

### Standardising the data.

Applying custom scale to only chosen columns, so that dummy variables are not scaled. 

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
```

```python
class CustomScaler (BaseEstimator, TransformerMixin):
    
    def __init__(self, columns):
        self.scaler = StandardScaler()
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y = None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y = None, copy = None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns = self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis = 1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
# Columns, which don't contain dummies. 
columns_to_scale = ['Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index','Children', 'Pets']
```

```python
# Declare absenteeism scaler.
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
# Calculate mean and standard deviation of each 'unscaled_input'.
absenteeism_scaler.fit(unscaled_inputs)
```

```python
# Substract mean and divide by standard deviation in order to scale the inputs.
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

### Creating the summary table with coefficients and intercept.

It's to use the model outside the Python. 

```python
# Intercept of the model.
reg.intercept_
```

```python
# Coefficients of the model.
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
# Creating the summary table with coefficients information.
summary_table = pd.DataFrame(columns = ['Feature name'], data = feature_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
# Adding intercept to the 'summary_table'.
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index(0)
summary_table
```

### Interpreting the coefficients/weights

```python
# Calculating the exponential value of coefficients/weights in 'Odds_ratio' column.
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending = False)
```

### Interpretation: 

Features are not particularily important if their coefficient is around 0 and odds ration around 1:
- Month Value 
- Daily Work Load Avearge
- Distance to Work 
- Day of the Week

We have 4 Reasons for absence. While creating the dummy variables for the Reasons, the Reason_0 was dropped to avoid multicollinearity. Reason_0 represented a situation where the person didn't give any particular reason for absence. Hence, the base model is when there is no reason for absence. Looking at the coefficients of Reasons 1-4, it can be said that there is a higher chance of excessive absence when there is a reason stated. 
E.g. It seems that people who report Reason_3 (poisoning) seems to be 22 times more likely to have excessive absence than people who were absent for no given reason (Reason_0).
